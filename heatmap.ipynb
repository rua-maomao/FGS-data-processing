```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path = '/content/HOT COMP.xlsx'
data = pd.read_excel(file_path)

# Generate correlation matrix
df_cleaned = data.apply(pd.to_numeric, errors='coerce')
df_cleaned = df_cleaned.dropna()
correlation_matrix = df_cleaned.corr()

# Set font size
sns.set(font_scale=1.4)

# Visualize
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, cbar_kws={'label': ''})

# Manually set the axis labels, ensuring the correct subscripts
plt.gca().set_xticklabels([
    "$\\Delta Z/Z_0$ at 1kHz",
    "Sunlight (mmol)",
    "Soil humidity (%)",
    "Temperature (℃)",
    "Soil EC (μs/cm)",
    "Relative humidity (%)"
], rotation=90)

plt.gca().set_yticklabels([
    "$\\Delta Z/Z_0$ at 1kHz",
    "Sunlight (mmol)",
    "Soil humidity (%)",
    "Temperature (℃)",
    "Soil EC (μs/cm)",
    "Relative humidity (%)"
], rotation=0)

plt.gca().xaxis.set_ticks_position('top')

plt.xticks(rotation=30)

plt.show()
```
