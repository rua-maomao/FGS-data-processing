```python
import math
def color_diff_percentage(rgb1, rgb2):
  distance = math.sqrt((rgb1[0] - rgb2[0])**2 +
   (rgb1[1] - rgb2[1])**2 +
    (rgb1[2] - rgb2[2])**2)
  max_distance = math.sqrt((255**2)*3)
  return (distance / max_distance) * 100
color1 = (81, 108, 26)
color2 = (78, 101, 27)
difference = color_diff_percentage(color1, color2)
print(f"difference:{difference:.2f}%")
```
